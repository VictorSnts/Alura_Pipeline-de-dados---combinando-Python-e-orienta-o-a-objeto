## Especificando versao da API

Versao Utilizada: 2022-11-28

```python
headers = {'X-GitHub-Api-Version': '2022-11-28'}
```

```python
r = requests.get('https://api.github.com/events', headers=headers)
```

Realizando uma requisição GET à API do Github para obter as informações do perfil de um usuário específico. 
Para isso, utilize a biblioteca Requests do Python e a URL https://api.github.com/users/{username}, substituindo {username} pelo nome de usuário desejado.

```python
username = "victorsnts"
r = requests.get(f'https://api.github.com/users/{username}', headers=headers)
print(f"Status Code: {r.status_code} \nURL: {r.url}")
```

Imprimindo na tela as seguintes informações:

Nome
Nome de usuário
Número de repositórios públicos
Data de criação da conta no GitHub

```python
print(f"Nome: {r.json()['name']}")
print(f"Nome de usuário: {r.json()['login']}")
print(f"Número de repositórios públicos: {r.json()['public_repos']}")
print(f"Data de criação da conta no GitHub: {r.json()['created_at']}")
```
